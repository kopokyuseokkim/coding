```python
import requests
import time
import datetime
import hmac
import hashlib
import random
import string

def send_solapi_final(text, to_number):
    # 1. 정보 입력 (사장님 정보를 꼭 넣어주세요)
    api_key = 'NCSBSYIUIGZ5YIW2'
    api_secret = 'KPC8UD7MSEETMCYO7CND8U31PDZOLC5E'
    from_number = '01024763226' # 솔라피에 등록된 발신번호 (숫자만)

    # 2. 솔라피 인증 헤더 생성
    now = datetime.datetime.now(datetime.timezone.utc).strftime('%Y-%m-%dT%H:%M:%S.%f')[:-3] + 'Z'
    salt = ''.join(random.choices(string.ascii_letters + string.digits, k=16))

    data = now + salt
    signature = hmac.new(
        api_secret.encode('utf-8'),
        data.encode('utf-8'),
        hashlib.sha256
    ).hexdigest()

    auth_header = f"HMAC-SHA256 apiKey={api_key}, date={now}, salt={salt}, signature={signature}"

    headers = {
        'Authorization': auth_header,
        'Content-Type': 'application/json'
    }

    url = "https://api.solapi.com/messages/v4/list"
    url = "https://api.solapi.com/messages/v4/send"

    payload = {
        'message': {
            'to': to_number,
            'from': from_number,
            'text': text
        }
    }

    # 4. 발송 요청
    try:
        response = requests.post(url, json=payload, headers=headers)

        if response.status_code == 200:
            print("✅ [성공] 드디어 문자가 발송되었습니다!")
            return response.json()
        else:
            print(f"❌ [실패] 상태 코드: {response.status_code}")
            print(f"오류 메시지: {response.text}")
    except Exception as e:
        print(f"❌ [예외]: {e}")
```
